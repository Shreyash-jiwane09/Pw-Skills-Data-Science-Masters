## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions.

**Flask is a Python web development framework that has many advantages, including:**

1. Ease of use :
Flask is lightweight and easy to get started with, making it suitable for beginners and intermediate-level Python users. It has a built-in development server and debugger, as well as excellent documentation.

2. Speed:
Flask can help developers create apps and services faster and more efficiently. It has integrated unit testing tools that provide immediate feedback on code, allowing developers to move quickly and avoid time-consuming bug fixes.

3. Flexibility :
Flask gives developers the freedom to experiment with their modules and architecture. It also allows developers to separate application logic from the presentation layer, which can lead to cleaner and more maintainable code.

4. Scalability:
Flask's microframework design makes it a good choice for projects that start small but have the potential to grow quickly.

5. Customization :
Flask comes with a built-in admin framework and other features that allow developers to create multilingual websites and add functionality like uploading, form validation, and object-relational mappers.


## Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_wolrd():
    return "hello_wolrd!"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

## Q3. What is App routing in Flask? Why do we use app routes?

App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 

Example: In our application, the URL (“/”) is associated with the root URL. So if our site’s domain was www.example.org and we want to add routing to “www.example.org/hello”, we would use “/hello”. 

To bind a function to an URL path we use the app.route decorator. In the below example, we have implemented the above routing in the flask.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
## route to show the following details:
### Company Name: ABC Corporation
### Location: India
###       Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def company_details():
    return "Company Name: ABC Corporation \n Location: India\nContact Detail: 999-999-9999¶"
    
@app.route("/welcome")
def welcome_msg():
    return "Welcome to ABC Corporation"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
