## Q1. What is an API? Give an example, where an API is used in real life.

API full form is an Application Programming Interface that is a collection of communication protocols and subroutines used by various programs to communicate between them. A programmer can make use of various API tools to make their program easier and simpler. Also, an API facilitates programmers with an efficient way to develop their software programs. Thus api meaning is when an API helps two programs or applications to communicate with each other by providing them with the necessary tools and functions. It takes the request from the user and sends it to the service provider and then again sends the result generated from the service provider to the desired user. 

A developer extensively uses APIs in his software to implement various features by using an API call without writing complex codes for the same. We can create an API for an operating system, database system, hardware system, JavaScript file, or similar object-oriented files. Also, an API is similar to a GUI(Graphical User Interface) with one major difference. Unlike GUIs, an application program interface helps software developers to access web tools while a GUI helps to make a program easier to understand for users.

## Q2. Give advantages and disadvantages of using API.

**Application Programming Interfaces (APIs) can offer many advantages, including:** 

1. Improved marketing : 
APIs can help companies expand their market reach and speed up the process of creating and marketing new products and services.
2. Simplified app development :
APIs can save time and money when developing apps, designing new products, and more.
3. Automation :
APIs can help automate tasks, which can increase efficiency, improve quality, and reduce risk.
4. Efficiency :
APIs can streamline processes by allowing different systems and platforms to share information and work together.
5. Improved security :
When websites use APIs, they are not immediately connected to a server when a request is submitted, which can reduce the risk of data breaches.
6. Adaptation :
APIs can help companies anticipate changes over time, which can make service provision more flexible.
7. Agility: 
APIs can help business systems integrate with other applications and processes, which can lead to smoother process automation.
8. Improved test coverage :
Automated API testing can reduce testing costs and make it easier to maintain tests and release products faster.


**Disadvantages of Application Programming  Interface (API) :**

1. Cost : 
Developing and maintaining APIs can be expensive in terms of time, support, and ongoing maintenance.
2. Complexity :
APIs can be complex and require a significant amount of development effort.
3. Compatibility :
APIs can have compatibility issues between different versions of the same software system or between different systems, which can lead to errors and system failures.
4. Security :
APIs can be vulnerable to a number of attacks, including man-in-the-middle, CSRF, XSS, SQL injection, and DDoS attacks. APIs can also be a hacker's primary target because they are a single point of entry.
5. Third-party dependencies :
APIs rely on third-party developers to create applications that integrate with the software system, which can create dependencies that are outside of the company's control. This can introduce reliability and performance issues if the API experiences downtime or changes its functionality.

## Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

### REST Architecture: 
REST stands for REpresentational State Transfer and API stands for Application Program Interface. REST is a software architectural style that defines the set of rules to be used for creating web services. Web services which follow the REST architectural style are known as RESTful web services. It allows requesting systems to access and manipulate web resources by using a uniform and predefined set of rules. Interaction in REST based systems happen through Internets Hypertext Transfer Protocol (HTTP). 

A Restful system consists of a:

1. client who requests for the resources.
2. server who has the resources.

**Architectural Constraints of RESTful API:** 
There are six architectural constraints which makes any web service are listed below:
1. Uniform Interface
2. Stateless
3. Cacheable
4. Client-Server
5. Layered System
6. Code on Demand


### SOAP Architecture : 

Simple Object Access Protocol(SOAP) is a network protocol for exchanging structured data between nodes. It uses XML format to transfer messages. It works on top of application layer protocols like HTTP and SMTP for notations and transmission. SOAP allows processes to communicate throughout platforms, languages, and operating system, since protocols like HTTP are already installed on all platforms. SOAP was designed by Bob Atkinson, Don Box, Dave Winer, and Mohsen Al-Ghosein at Microsoft in 1998. SOAP was maintained by the XML Protocol Working Group of the World Wide Web Consortium until 2009.
