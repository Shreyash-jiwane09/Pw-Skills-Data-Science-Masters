## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Below are the five plots that we can plot using Seaborn library along with their usage :
    1. Line Plot 
    2. Count Plot 
    3. Bar Chart
    4. Pair plot 
    5. Scatter Plot 

**1. Line Plot :** A Line Plot Is The Simplest Plot In All Plotting Types, As It Is The Visualization Of A Single Function. This Plot Helps Us To See The Relationship Between X-Axis, Y-Axis And It Also Takes Some Parameters Such As Hue, Size, Color, Etc.  

 

#### Code: 

```python
import seaborn as sns
x = [2,4,6]
y = [1,2,3]
sns.lineplot(x)
```

**2. Count Plot :**
A Count Plot Is Used To Show The Counts Of Observations In Each Categorical Bin Using Bars.
This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding. Data Is Also An Optional Parameter That Takes DataFrame, Array, Or List Of Arrays, Dataset For Plotting.

#### Code: 

```python
import seaborn as sns
iris=sns.load_dataset("iris")
sns.countplot(x="species",data = iris)
```

**3. Bar Chart :** A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. The Bar Chart Displays Data.Using Several Bars, Each Representing A Particular Category. This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding.

### Code: 

```python
import seaborn as sns
sns.barplot(x=iris.species,y=iris.sepal_length)
```

**4. Pair plot :**
Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column. The Diagonal Plots Are A Univariate Distribution Plot That Helps To Draw The Marginal Distribution Of The Data In Each Column.
A Pair Plot Pairwise Relationships With Other Columns In The Data Frame And Also Plot Pair Plot With Itself.

### Code :


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(data=iris,hue="species")
plt.show()
```

**5. Scatter Plot :** Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape. The Position Of Each Marker On The Horizontal And Vertical Axis Indicates Values For An Individual Data Point. This Plot Is Used To Observe Relationships Between Variables.
This Method Is Accepting The Parameters X, Y  Which Take The Name Of A Variable In Data, Hue It Is An Optional Parameter It Helps To Take Column Name For Color Encoding. Marker Is A Parameter That Is Used To Change The Shape Of The Point Which Is Shown In The Plot.

### Code: 

```python
sns.scatterplot(x=iris.sepal_width,y=iris.sepal_width)
```

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

## Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
#imporing seaborn and matplotlib
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#loading dataset 
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.value_counts("event")
```

```python
fmri.value_counts("region")
```

```python
##Using sns.lineplot to plot line plot
plt.figure(figsize=(8,4))
sns.lineplot(data=fmri,x='timepoint',y='signal',style='region',hue='event')
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
##loading "Flights" dataset 
flights=sns.load_dataset("flights")
flights.head()
```

### Pivot the dataset to create a matrix with month on the x-axis, year on the y-axis, and passengers as values


```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
##plotting heatmap
import matplotlib.pyplot as plt
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True, 
            fmt="d", 
            linewidths=1.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
